```python
#2.	Напишите программу, которая принимает строку от пользователя и выводит количество символов в этой строке без учета пробелов.
a2 = input ('Введите строку ')
i = 0
for symbol in a2:
  if symbol != " ":
    i = i + 1
print ('Количество символов без пробелов = ', i)
```
